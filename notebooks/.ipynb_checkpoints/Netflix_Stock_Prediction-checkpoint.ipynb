```python
#Change date from object to date/time & Setting date as index
netflix['Date'] = pd.to_datetime(netflix['Date'])
netflix = netflix.sort_values('Date')
netflix.set_index('Date', inplace=True)
netflix.head()
```

```python
plt.figure(figsize=(9,3))
plt.plot(netflix['Volume'])
plt.title('Netflix Trading Volume Over Time')
plt.xlabel('Year')
plt.ylabel('Volume')
plt.show()

# Insight --> Trading Volume Behavior
# --> Trading volume spikes during major events (e.g. earnings announcements, stock drops).
# --> You may notice higher trading volume during price crashes  showing panic selling.
# --> After large crashes, volume usually drops back to normal levels.
```

```python
# Create moving average features
netflix['Moving_Avg10'] = netflix['Close'].rolling(window=10).mean()
netflix['Moving_Avg20'] = netflix['Close'].rolling(window=20).mean()
netflix['Moving_Avg50'] = netflix['Close'].rolling(window=50).mean()
netflix['Moving_Avg100'] = netflix['Close'].rolling(window=100).mean()

netflix.tail()
```

```python
# Predict next day's closing price
netflix['Target'] = netflix['Close'].shift(-1)

# Remove last row --> no target
netflix = netflix[:-1]

netflix.tail()
```

```python
features = ['Close', 'Moving_Avg10', 'Moving_Avg20', 'Moving_Avg50', 'Moving_Avg100']
X = netflix[features]
y = netflix['Target']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False)  # time series  no shuffle!
```

```python
# Create daily returns
netflix['Return'] = netflix['Close'].pct_change()

# Create lagged features (previous close prices)
netflix['Lag1'] = netflix['Close'].shift(1)
netflix['Lag2'] = netflix['Close'].shift(2)
netflix['Lag3'] = netflix['Close'].shift(3)

# Remove missing values due to shifting
netflix = netflix.dropna()

netflix.head()
```

```python
# Remove any remaining NaN rows in features or target
netflix = netflix.dropna()

# X and y after cleaning
features = ['Close', 'Moving_Avg10', 'Moving_Avg20', 'Moving_Avg50', 'Moving_Avg100', 'Return', 'Lag1', 'Lag2', 'Lag3']
X = netflix[features]
y = netflix['Target']

# Re-split into training and testing (again)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)
```

```python
 netflix['Target']
```

```python
# Initialize model
model_lr = LinearRegression()

# Train model
model_lr.fit(X_train, y_train)
# Predict on test data
y_pred_lr = model_lr.predict(X_test)

# Evaluate model
mse = mean_squared_error(y_test, y_pred_lr)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred_lr)

mse, rmse, r2
```

```python
# Train model
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)

# Calculate residuals
residuals = y_test - y_pred

# Plot residuals over time
plt.figure(figsize=(9,3))
plt.plot(y_test.index, residuals)
plt.title("Residuals Over Time (Linear Regression)")
plt.xlabel("Date")
plt.ylabel("Prediction Error (Residual)")
plt.axhline(0)
plt.show()
```

```python
# Step 1: Define X and y
X = netflix[['Close', 'Moving_Avg10', 'Moving_Avg20', 'Moving_Avg50', 'Moving_Avg100', 'Return', 'Lag1', 'Lag2', 'Lag3']]
y = netflix['Target']

# Step 2: Train-Test Split (keep time order)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)

# Step 3: Train Linear Regression Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)

# Step 4: Predict
y_pred = model.predict(X_test)

# Step 5: Plot Actual vs Predicted
import matplotlib.pyplot as plt

test_dates = y_test.index  # Use date index for alignment

plt.figure(figsize=(9,3))
plt.plot(test_dates, y_test, label="Actual", linewidth=2)
plt.plot(test_dates, y_pred, label="Predicted", linestyle='dashed')
plt.xlabel("Date")
plt.ylabel("Target")
plt.title("Actual vs Predicted Trend (Linear Regression)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Calculate metrics
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

# Display results
print("Model Performance:")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (R): {r2:.4f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Feature list
features = ['Close', 'Moving_Avg10', 'Moving_Avg20', 'Moving_Avg50', 
            'Moving_Avg100', 'Return', 'Lag1', 'Lag2', 'Lag3']

# Get feature importance from trained XGBoost model
importances = xgb_model.feature_importances_

# Sort the importances for better visualization
indices = np.argsort(importances)

# Plot feature importance
plt.figure(figsize=(10, 6))
plt.title("Feature Importance - XGBoost")
plt.barh(range(len(indices)), importances[indices], align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel("Importance Score")
plt.ylabel("Features")
plt.show()
```

```python
# Create a feature importance table
feature_importance_table = sorted(zip(features, xgb_model.feature_importances_), key=lambda x: x[1], reverse=True)

print("Feature Importance Ranking (XGBoost):")
print("-------------------------------------")
for rank, (feature, importance) in enumerate(feature_importance_table, start=1):
    print(f"{rank}. {feature}: {importance:.4f}")
```

```python
# Create train-test split for Holt-Winters
# (Using only 'Target' since Holt-Winters is univariate)
train_size = int(len(netflix) * 0.8)
train = netflix['Target'][:train_size]
test = netflix['Target'][train_size:]
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Holt-Winters model (multiplicative trend and seasonality)
holt_model = ExponentialSmoothing(
    train,
    trend="mul",
    seasonal=None,  # If no seasonality, set to None
).fit()

# Forecast for test period
y_pred_hw = holt_model.forecast(len(test))
```

```python
plt.figure(figsize=(9,3))

# Plot full actual closing price
plt.plot(netflix.index, netflix['Target'], label="Actual Price", linewidth=2)

# Plot Linear Regression predictions aligned to test index
plt.plot(y_test.index, y_pred_lr, label="Linear Regression", linestyle='--')

# Plot Holt-Winters predictions aligned to test index
plt.plot(y_test.index, y_pred_hw, label="Holt-Winters", linestyle='--')

# Plot XGBoost predictions aligned to test index
plt.plot(y_test.index, y_pred_xgb, label="XGBoost", linestyle='--')

plt.title("Actual vs Predicted Stock Prices (Full Timeline Comparison)")
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Evaluate Holt-Winters
mae_hw = mean_absolute_error(test, y_pred_hw)
rmse_hw = np.sqrt(mean_squared_error(test, y_pred_hw))
r2_hw = r2_score(test, y_pred_hw)

print("Holt-Winters Performance:")
print("MAE:", mae_hw)
print("RMSE:", rmse_hw)
print("R Score:", r2_hw)
```
